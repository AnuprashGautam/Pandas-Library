# Important functions and attributes

## head()
- It gives the preview of the data.
- By default it print the first 5 rows.

```python
import pandas as pd
data=pd.read_csv('matches.csv')
data.head(10)
```

## tail()
- This function print the last rows.
- By default it print the last 5 rows.

```python
import pandas as pd
data=pd.read_csv('matches.csv')
data.tail(10)
```

## shape
- It is an attribute that gives the shape of the data.
- It return a tuple having 2 elements. First--> No of Rows and Second--> No of Columns

```python
print(data.shape)
print(data.shape[0])
print(data.shape[1])
```
