# Fetching Rows and Columns using iloc 

## Fetching Column(s)
- We can either use double quotes or single quotes.
- For single column directly we can pass the quoted column name and for multiple columns we have to pass the list composing the quoted columns' name.
- Single column--> Series
- Multiple Columns--> Dataframe

```python
import pandas as pd 
data=pd.read_csv('matches.csv')
print(data["winner"])
print()
print(type(data["winner"]))
print()
print(data["winner"].shape)

```

```python
print(data[["team1","team2","winner"]])
print()
print(type(data[["team1","team2","winner"]]))
print()
print(data[["team1","team2","winner"]].shape)
```

```python
type(data[["team1","team2","winner"]])
```

## Fetching Row(s) using iloc() function
- For that iloc() function is used. *iloc* means index location.
- We can also access the multiple rows by inputing the index numbers separrated by comma.
- Because indexing start with the 0 so for 0 it returns the first row.
- For -1 it return the last row.
- We can also apply the slicing concept here.

```python
data.iloc[0]
```

```python
data.iloc[[2,3,5]]
```

```python
data.iloc[-1]
```

```python
data.iloc[::2]
```

## NoteðŸ””
- We can use iloc function to access the columns too.
- Portion before comma, is used for the *rows* and portion after the comma is used for the *columns*.

```python
data.iloc[:,[3,5,7]]
```
