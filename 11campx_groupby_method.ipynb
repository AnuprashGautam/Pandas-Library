# goupby() method

- This function is used to group a categorical-column data into groups.

```python
import pandas as pd

company=pd.read_csv("Fortune 500 Companies.csv")
company.head(2)
```

```python
company.groupby("sector")
```

```python
sectors=company.groupby("sector")
sectors
```

```python
len(sectors)
```

```python
len(company)
```

```python
sectors.size()
```

```python
sectors.size().sort_values(ascending=False)
```

```python
sectors.first()      #Return the first dataline of each sector.
```

```python
sectors.last()      #Return the last dataline of each sector.
```

## groups 
- It is an attribute that return a dictionary.
- Key contain the name of the group.
- Value contain the ids of all the elements of the group contain.

```python
sectors.groups
company.iloc[23,:]
```

## get_group()
- Give the details of the mentioned group.


```python
sectors.get_group("Energy")
```

## sum()
- Add the whole column within the same group and print all the details of the group.

### Note: Similarly we can calulate the mean by the mean() function.

```python
sectors.sum()
```

```python
sectors["year"].mean()
```
